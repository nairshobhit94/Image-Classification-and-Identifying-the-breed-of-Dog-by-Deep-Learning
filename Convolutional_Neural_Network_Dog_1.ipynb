```python
import numpy as np

from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from keras.applications.resnet50 import preprocess_input, decode_predictions
```

# Load Image

```python
img=load_img("dog_4.jpg", target_size= (336,336))
```

```python
img
```

```python
from keras.applications.resnet50 import ResNet50
```

# ResNet-50 is a convolutional neural network that is trained on more than a million images from the ImageNet database

```python
model=ResNet50(weights="imagenet")
```

# Reduction of Size of the Image

```python
img=load_img("dog_4.jpg", target_size = (224, 224))
img
```

```python
img=img_to_array(img)
print(img.shape)
```

```python
img=np.expand_dims(img, axis=0)
print(img.shape)
```

```python
## prediction without preprocessing
pred_class=model.predict(img)
```

```python
print(pred_class)
```

```python
# print only top 10 predicted classes
n=10
top_n=decode_predictions(pred_class, top=n)
```

```python
for c in top_n[0]:
    print(c)
```

```python
img=preprocess_input(img)
print(img.shape)
```

```python
## prediction with preprocessing
pred_class=model.predict(img)
# print(pred_class)
```

```python
n=10
top_n=decode_predictions(pred_class, top=n)
```

# Predicting the Breed of The Dog

```python
for c in top_n[0]:
    print(c)
```
